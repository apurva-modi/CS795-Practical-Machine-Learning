# Introduction to Matplotlib
Matplotlib is a popular plotting library in python. Matplotlib allows one to create many different types of plots.
Detail can be found at
https://matplotlib.org/contents.html

## Import plt library
matplotlib.pyplot includes the methods of creating different kinds of plots.
%matplotlib inline is a jupyter notebook specific command that allow one to display plots in the notebook.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Create a simple scatter plot
Without specifying x-axis values, plt will start from 0 to as many as the number of items in y-axis values with step size 1.

```python
plt.plot([1, 2, 4, 8, 16])
```

## Specifying explicit x-axis values in the simple scatter plot
plt.show() removes the [<matplotlib.lines.Line2D at 0xfcb14e0>] line.

```python
plt.plot([-2, -1, 1, 2, 3], [1, 2, 4, 8, 16])
plt.show()
```

## Scatter plot with format arguments
plt.plot accepts (x, y, format) as arguments.

The "format" argument is a combination of {color}{marker}{line}. 

For example, "ro-" means red, circle, with line. "b*--" means blue, star, with dashed line.

```python
plt.plot([-2, -1, 1, 2, 3], [1, 2, 4, 8, 16], "ro-")
plt.show()
```

## Draw two sets of points

```python
plt.plot([-2, -1, 1, 2, 3], [1, 2, 4, 8, 16], "ro-")
plt.plot([-4, -2, 0, 2, 4], [2, 3, 1, 4, 6], "b*--")
plt.show()
```

## Adding Title, Legend, X and Y axis labels

```python
plt.plot([-2, -1, 1, 2, 3], [1, 2, 4, 8, 16], "ro-", label="red data")
plt.plot([-4, -2, 0, 2, 4], [2, 3, 1, 4, 6], "b*--", label="blue data")
plt.title('Scatterplot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='best') # best location
plt.show()
```

## Drawing a function

```python
import numpy as np
X = np.linspace(0,2*np.pi,1000)
plt.plot(X,np.sin(X))
plt.plot(X,np.cos(X))
plt.xlabel("X")
plt.show()
```

## Scatter plot with variable dot size and color
plt.scatter() allows one to plot a scatter plot with variable dot size and color

```python
x = np.random.rand(100)
y = np.random.rand(100)
color = np.random.rand(100)*255
size = np.random.rand(100)*500
plt.scatter(x, y, c=color, s=size)
plt.title("Scatter Plot")
plt.colorbar()
plt.show()
```

## Step plot

```python
n = np.array([1, 2, 3, 4, 5])
plt.step(n, n**2, lw=2)
plt.title("Step Plot")
plt.show()
```

## Bar chart

```python
n = np.array([1, 2, 3, 4, 5])
plt.bar(n, n**2, align="center", width=0.5, alpha=0.5)
plt.title("Bar Chart")
plt.show()
```

## Fill between plot

```python
x = np.linspace(0, 5, 10)
plt.fill_between(x, x**0.5, x**2, color="steelblue", alpha=0.5)
plt.title("Fill Between")
plt.show()
```

## Histogram

```python
x = np.linspace(0, 5, 100)
bins = [-3, -2, -1, 0, 1, 2, 3, 4, 5, 6, 7]
plt.hist(x + np.random.randn(len(x)), bins)
plt.title("Histogram")
plt.show()
```

## Box plot
The box extends from the lower to upper quartile values of the data, with a line at the median. The whiskers extend from the box to show the range of the data.

```python
x = np.random.randn(1000, 2)
x[:, 1] *= 3
plt.boxplot(x)
plt.title("Box Plot")
plt.show()
```

## plotting multiple figures together

```python
plt.subplot(1,2,1)  # (nRows, nColumns, axes number to plot)
plt.plot([1,2,3,4,5], [1,2,3,4,10], 'go')  # green dots
plt.title('Scatterplot Greendots')  
plt.xlabel('X'); plt.ylabel('Y')
plt.xlim(0, 6); plt.ylim(0, 12)

# Right hand side plot
plt.subplot(1,2,2)
plt.plot([1,2,3,4,5], [2,3,4,5,11], 'b*')  # blue stars
plt.title('Scatterplot Bluestars')  
plt.xlabel('X'); plt.ylabel('Y')
plt.xlim(0, 6); plt.ylim(0, 12)
plt.show()
```

```python
plt.subplot(2,1,1)  # (nRows, nColumns, axes number to plot)
plt.plot([1,2,3,4,5], [1,2,3,4,10], 'go')  # green dots
plt.title('Scatterplots Greendots and Bluestars')  
plt.xlabel('X'); plt.ylabel('Y')
plt.xlim(0, 6); plt.ylim(0, 12)

# Right hand side plot
plt.subplot(2,1,2)
plt.plot([1,2,3,4,5], [2,3,4,5,11], 'b*')  # blue stars
plt.xlabel('X'); plt.ylabel('Y')
plt.xlim(0, 6); plt.ylim(0, 12)
plt.show()
```
